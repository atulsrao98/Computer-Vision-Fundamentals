```python
import cv2
```

```python
def draw_bbox(img,bboxes,color=(0,0,255),thick=3):
    copy_img=np.copy(img)
    x1=bboxes[0][0][0]
    y1=bboxes[0][0][1]
    x2=bboxes[0][1][0]
    y2=bboxes[0][1][1]
    x3=bboxes[1][0][0]
    y3=bboxes[1][0][1]
    x4=bboxes[1][1][0]
    y4=bboxes[1][1][1]
    x5=bboxes[2][0][0]
    y5=bboxes[2][0][1]
    x6=bboxes[2][1][0]
    y6=bboxes[2][1][1]
    cv2.rectangle(copy_img,(x1,y1),(x2,y2),color,thick)
    cv2.rectangle(copy_img,(x3,y3),(x4,y4),color,thick)
    cv2.rectangle(copy_img,(x5,y5),(x6,y6),color,thick)
    return copy_img
```

```python
bboxes=[((275, 500), (400, 600)), ((840, 500), (1200, 700)),((600,510),(650,550))]
```

```python
result=draw_bbox(img,bboxes)
plt.imshow(result)
```
