```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as img
```

```python
image=img.imread(r"D:\Downloads\canny-test.jpg")
plt.imshow(image)
```

```python
import cv2
bw_image=cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(bw_image,cmap='gray')
```

```python
kernelsize=3
blur_bw=cv2.GaussianBlur(bw_image,(kernelsize,kernelsize),0)
```

```python
plt.imshow(blur_bw,cmap='gray')
```

```python
low_threshold=92
high_threshold=120
```

```python
edges=cv2.Canny(blur_bw,low_threshold,high_threshold)
```

```python
plt.imshow(edges,cmap='Greys_r')
```
